# Zadanie 2.1 – Pierwszy layout turbin w TopFarm2

Dzisiaj po prostu rysujemy punkty (turbiny) na płaszczyźnie x-y.

```python
import numpy as np
import matplotlib.pyplot as plt

# test czy TopFarm2 się zaimportowało
import topfarm
print("TopFarm2 wersja:", topfarm.__version__)

# przykładowe położenie 4 turbin (w metrach)
x = np.array([0, 500, 1000, 1500])
y = np.array([0, 0, 500, 500])

# kolory dla każdej turbiny
colors = ['darkblue', 'green', 'red', 'orange']
labels = ['T1', 'T2', 'T3', 'T4']

plt.figure(figsize=(10, 7))

# rysowanie punktów z różnymi kolorami
for i, (xi, yi, color, label) in enumerate(zip(x, y, colors, labels)):
    plt.scatter(xi, yi, s=200, color=color, marker='o', alpha=0.7, edgecolors='black', linewidth=2)
    # dodanie etykiet tekstowych
    plt.annotate(label, (xi, yi), textcoords="offset points", xytext=(0, 15), 
                 ha='center', fontsize=12, fontweight='bold', color=color)

# dodanie strzałek między turbinami w formie połączeń
for i in range(len(x) - 1):
    plt.annotate('', xy=(x[i+1], y[i+1]), xytext=(x[i], y[i]),
                arrowprops=dict(arrowstyle='->', lw=1.5, color='gray', alpha=0.5))

plt.title("Layout turbin – Zadanie 2.1 (z etykietami)", fontsize=14, fontweight='bold')
plt.xlabel("x [m]", fontsize=12)
plt.ylabel("y [m]", fontsize=12)
plt.grid(True, alpha=0.3)
plt.axis("equal")           # ważne – żeby nie zniekształcać odległości
plt.legend(['T1', 'T2', 'T3', 'T4'], loc='upper right')
plt.tight_layout()
plt.show()
```

## Co zostało dodane?

W tym zadaniu ulepszono podstawową wizualizację layout'u turbin poprzez:

1. **Indywidualne kolory dla każdej turbiny** – zamiast jednolitego koloru, każda turbina ma przypisany inny kolor (niebieski, zielony, czerwony, pomarańczowy), co ułatwia ich identyfikację na wykresie.

2. **Etykiety tekstowe** – każdy punkt jest opisany unikatową etykietą (T1, T2, T3, T4), która jest wyświetlana nad każdą turbiną z odpowiadającym jej kolorem.

3. **Strzałki połączeń** – dodano graficzną reprezentację połączeń między turbinami w sekwencji, co sugeruje możliwą trasę lub hierarchię.

4. **Ulepszony styl** – zwiększono rozmiar punktów, dodano czarne obramowanie dla kontrastu, oraz ulepszone etykiety osi i tytuł wykresu.

## Czego nauczyliśmy się w tym zadaniu?

To zadanie pokazało nam praktyczne zastosowanie bibliotek Python do wizualizacji danych przestrzennych:

- **matplotlib.scatter()** – sposób rysowania punktów na płaszczyźnie xy
- **Pętle i iteracje** – rysowanie wielu elementów o różnych atrybutach (kolory, etykiety)
- **plt.annotate()** – dodawanie tekstu i grafiki (strzałek) do wykresu
- **Parametry wykresu** – używanie `axis('equal')` do zachowania proporcji, `grid()` do lepszej czytelności
- **TopFarm2** – wstęp do biblioteki dedykowanej optymalizacji farm wiatrowych

Umiejętności te będą fundamentem do bardziej zaawansowanych analiz rozmieszczenia turbin i optymalizacji ich pozycji.
